```python
#For checking missing values
df.isnull().sum()

# We look into the missing value column 'Review Ratng' 
# How should we replace the null value 'mean','median'.
# Mean can get affected by outliers and that can skew are data.
# Therefore, We are choosing median because it is robust to outliers.
```

```python
df['Category'].unique()
```

```python
# In this we will calculate median for each category(cloth, footwear...).
# Because ratings can vary a lot depending on each category.
#eg: cloth can have a different rating than footwear.
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))

#'groupby' we split data into group based on category.
#'Review Rating' focuses only on Review rating column within category.
#'transform(..)' Applies a calculation and keeps the original dataframe shape.
#x.fillna(x.median()) for each category replaces missing (NaN), 
#review ratings with the median rating of that category.
```

```python
#for checking purpose only for missing value.
df.isnull().sum()
```

```python
# column names problem (uppercase, lowercase, spacess.. etc)
# So put all the column name in snake casing eg:[review_rating]
# by this it is easier to read and to reference the code SQL or Python without running into errors.
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})

```

```python
df.columns
```

```python
# feature engineering for some new column
# create a column age_group

labels = ['Young Adult','Adult', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)

#It divides the age column into 4 equal-sized groups.
#df['age'] - the age data.
#q=4 - split the data into 4 groups (quartiles).
#labels=l - give each group a name.
```

```python
df[['age', 'age_group']].head(10)
```

```python
#creating column purchase_frequency_days

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# Do we need both of these column lets check.
#'all()' it verifies weather every row in the Dataframe has the same value in both columns.
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# We don't need both the columns because both have the same values.
# let remove promo_code_used.
df = df.drop('promo_code_used', axis=1)
```
